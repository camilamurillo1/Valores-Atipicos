```python
#iltroNum=data.iloc[ : , 1:3]       #Columnas consecutivas 
filtroNum=data.iloc[ : , [8,9,10, 11, 12, 13, 14]]  #Columnas no consecutivas 
filtroNum
```

```python
filtroLetra=data.iloc[ : , [1,2,3,4,5,6,7,15,16,17]]  #Columnas no consecutivas 
filtroLetra
```

**PROCEDIMIENTO PARA ELIMINAR VALORES NULOS EN DATAFRAME**

```python
#Reemplazamos valores nulos del dataframe con "bfill"
data1 =data.fillna(method="bfill") 
data1
```

```python
#Corroboramos valores nulos del dataframe
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores nulos del dataframe con "mean"
#Realizamos una copia del dataframe
data2=data1.copy()
data2=data2.fillna(round(data.mean(),1))
data2
```

```python
data3 =data2.fillna(method="ffill") 
data3
```

```python
#Corroboramos valores nulos del dataframe
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#SOLO PARA FILTRO CUANTITATIVO 

#Realizamos diagrama de caja o bigote de cada columna del dataframe
fig = plt.figure(figsize =(15, 8))
filtroNum.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show() #dibujamos el diagrama
```

**PROCEDIMIENTO "DESVIACIÓN ESTÁNDAR" PARA ELIMINAR OUTLIERS EN DATAFRAME**

```python
#Método aplicando desviación estandar. Encuentro los valores extremos
y=filtroNum
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Encontramos Ouliers del Dataframe
outliers= filtroNum[(y>Limite_Superior)|(y<Limite_Inferior)]
outliers

#todos los que on ouliers son numericos, y los que no son insiders
```

```python
#Obtenemos datos y los outliers se convierten en nulos en el DataFrame
data4= filtroNum[(y<=Limite_Superior)&(y>=Limite_Inferior)]
data4
```

```python
#Eliminamos columnas innecesarias
#data4= data3.drop(['indice_tiempo'], axis=1)
#data4
```

```python
#Corroboramos valores nulos del dataframe4
valores_nulos=data4.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data_clean=data4.copy()
data_clean=data_clean.fillna(round(data4.mean(),1))
data_clean
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data_clean.isnull().sum()
```

**PROCEDIMIENTO "CUANTILES" PARA SUSTITUIR OUTLIERS EN DATAFRAME**

```python
#Método aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y=filtroNum

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Encontramos Outliers del Dataframe
outliers_iqr= filtroNum[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
#Obtenemos datos limpios del Dataframe
data3_iqr= filtroNum[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data3_iqr
```

```python
#Eliminamos columnas innecesarias
#data4_iqr= data3_iqr.drop(['indice_tiempo'], axis=1)
#data4_iqr
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data3_iqr.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data4_iqr=data3_iqr.copy()
data4_iqr=data4_iqr.fillna(round(data3_iqr.mean(),1))
data4_iqr
```

```python
# Unimos la columna 'indice_tiempo' con el dataframe

#indice_tiempo=data2['indice_tiempo']
Datos_limpiosD = pd.concat([data_clean, filtroLetra], axis=1)
Datos_limpiosD
```

```python
Datos_limpiosQ = pd.concat([data4_iqr,filtroLetra], axis=1)
Datos_limpiosQ
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=Datos_limpiosQ.isnull().sum()
valores_nulos
```
