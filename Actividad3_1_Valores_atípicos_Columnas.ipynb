```python
#Carga desde un archivo .csv sin indice
data= pd.read_csv('Ventas_totalesP.csv') 
```

```python
#Obtengo información del Dataframe
data.head()
```

```python
#Corroboramos valores nulos
valores_nulos=data.isnull().sum().sum()
valores_nulos
```

```python
#Realizamos histograma de variables cuantitativas
fig = plt.figure(figsize =(5, 3))
plt.hist(x=data["otros"], color='#D68910', rwidth=0.50)
plt.title('Histograma de otros')
plt.xlabel('Otros')
plt.ylabel('Frecuencia')

plt.show() #dibujamos el histograma
```

```python
#Realizamos diagrama de caja o bigote
fig = plt.figure(figsize =(5, 3))
plt.boxplot(data["otros"]) 
plt.title("Outliers de otros")
plt.show() #dibujamos el diagrama
```

**Método de desviación estándar**



```python
#Método aplicando desviación estandar. Encuentro los valores extremos
y=data["otros"]
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Encontramos Ouliers
outliers= data[(y>Limite_Superior)|(y<Limite_Inferior)]
outliers
```

```python
#Obtenemos datos limpios
data_clean= data[(y<=Limite_Superior)&(y>=Limite_Inferior)]
data_clean
```

```python
#Realizamos diagrama de caja o bigote
fig = plt.figure(figsize =(5, 3))
plt.boxplot(data_clean["otros"]) 
plt.title("Outliers de otros")
plt.show() #dibujamos el diagrama
```

**MÉTODO DE IQR (RANGO INTER QUARTILICO)**

```python
#Método aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y=data["otros"]

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Encontramos Ouliers
outliers_iqr= data[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
#Obtenemos datos limpios
data_clean_iqr= data[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data_clean_iqr

#Aqui ya borramos los datos nulos 
```

```python
#Realizamos diagrama de caja o bigote
fig = plt.figure(figsize =(5, 4.5))
plt.boxplot(data_clean_iqr["otros"]) 
plt.title("Outliers de otros")
plt.show() #dibujamos el diagrama
```

```python
valores_nulos=data_clean_iqr.isnull().sum().sum()
valores_nulos
```
